```python
import pandas as pd
import glob

files = glob.glob('/content/ce_data/*.parquet')

dfs = []

#randomly select 1000–3000 rows from each file
for f in files:
    temp = pd.read_parquet(f)
    dfs.append(temp.sample(1500, random_state=42))

df_sample = pd.concat(dfs, ignore_index=True)

df_sample.shape
```

```python
#identify categorical vs numerical features
categorical_cols = df_sample.select_dtypes(include=['object']).columns
numeric_cols = df_sample.select_dtypes(include=['int64', 'float64']).columns

categorical_cols, numeric_cols
```

```python
#check unique values of key categorical columns
for col in categorical_cols[:5]:
    print(col, df_sample[col].nunique())
```

```python
#quick correlation check between numeric features
df_sample[numeric_cols].corr().head()
```

```python
#drop columns that are completely missing
#check how many columns we have before cleaning
print("Number of columns before dropping all-NaN columns:", df_sample.shape[1])

#drop columns where **all** values are NaN
df_clean = df_sample.dropna(axis=1, how='all')

# check how many columns remain
print("Number of columns after dropping all-NaN columns:", df_clean.shape[1])
```

```python
#check missing value percentage for each column

#compute missing ratio for each column
missing_ratio = df_clean.isna().mean().sort_values(ascending=False)

#show top 20 columns with the most missing values
missing_ratio.head(20)
```

```python
# drop columns with >85% missing rate
missing_ratio = df_clean.isna().mean()
cols_to_drop = missing_ratio[missing_ratio > 0.85].index

df_clean2 = df_clean.drop(columns=cols_to_drop)

print("Numbers of dropped colomns：", len(cols_to_drop))
print("Numbers of col after cleaning：", df_clean2.shape[1])
```

```python
#fill missing values for categorical features

#to object
for col in df_clean2.columns:
    if df_clean2[col].dtype == "object":
        df_clean2[col] = df_clean2[col].fillna("Unknown")

#to numeric
for col in df_clean2.columns:
    if df_clean2[col].dtype != "object":
        df_clean2[col] = df_clean2[col].fillna(df_clean2[col].median())
```
